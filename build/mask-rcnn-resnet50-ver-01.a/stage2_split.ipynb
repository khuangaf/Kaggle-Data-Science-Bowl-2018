```python
def checkGray(img):
    threshold = 2
    H, W, _ = img.shape
    for h in range(H):
        for w in range(W):
            R = img[h, w,0]
            G = img[h, w, 1]
            B = img[h,w,2]
            if abs(R-G) > threshold or abs(R-B) > threshold or abs(B-G) > threshold:
                return False
    return True
```

```python
checkGray(img)
```

```python
fig, ax  = plt.subplots( figsize=(100,100))
ax.imshow(img)
```

```python
from tqdm import tqdm
```

```python
stage2_test_color = []
stage2_test_gray = []
for t_id in tqdm(stage2_test):
    path = '/data/steeve/DSB/data/__download__/stage2_test_final/' + t_id + '/images/' + t_id + '.png'
    img = cv2.imread(path)
    if checkGray(img):
        stage2_test_gray.append(t_id)
    else:
        stage2_test_color.append(t_id)
```
